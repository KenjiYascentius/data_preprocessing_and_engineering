### Label data hilang

```python
from numpy import nan
from pandas import read_csv

# data = read_csv('north.csv')
data = read_csv('north_20000.csv')

# Melihat jumlah missing value per kolom
num_missing = (data == -9999).sum()
print(num_missing)

data = data.replace(-9999,nan)
display(data.head(20))
```

Visualisasi distribusi data sebelum imputasi

```python
from matplotlib import pyplot as plt

def show_distribution(data):
    columns = data.columns
    fig = data.hist(xlabelsize=4,ylabelsize=4)
    [x.title.set_size(4) for x in fig.ravel()]
    plt.show()

show_distribution(data.iloc[:, 3:20])
```

### Imputasi Data

Menggunakan Iterative Imputer

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from numpy import nan
from pandas import read_csv

# data = read_csv('north.csv')
data = read_csv('north_20000.csv')

data = data.replace(-9999,nan)

# display(data.head(5))

# cek jumlah missing value sebelom di imputasi 
print(data.isnull().sum().sum())

# define imputer
imputer = IterativeImputer(n_nearest_features=None,imputation_order='ascending',max_iter=300) #Pasang max_iter=300 agar convergent
data.iloc[:, 3:20] = imputer.fit_transform(data.iloc[:, 3:20])

print(data.isnull().sum().sum())

display(data.head(5))

data_iterative = data 
```

Menggunakan KNN Imputer

```python
from numpy import nan
from pandas import read_csv
from sklearn.impute import KNNImputer

# data = read_csv('north.csv')
data = read_csv('north_20000.csv')

data = data.replace(-9999,nan)
# display(data.head(5))
print(data.isnull().sum().sum())

imputer = KNNImputer(n_neighbors=5)
data.iloc[:, 3:20] = imputer.fit_transform(data.iloc[:, 3:20])
print(data.isnull().sum().sum())
display(data.head(5))

data_knn = data
```

Distribusi data sebelum imputasi

## Kesimpulan


Dataset ini memiliki banyak data -9999. Karena -9999 merupakan nilai yang tidak masuk akal untuk kolom-kolom yang ada, dapat disimpulkan bahwa data-data tersebut adalah missing data. Karena jumlahnya banyak saya memilih untuk melakukan imputasi data daripada dibuang.

Sebelum missing data diimputasi dapat dilihat dari distribusi bahwa data sama sekali tidak dapat digunakan sehingga imputasi perlu dilakukan. Kita menggunakan dua jenis imputasi yaitu iterative dan KNN. Setelah dibandingkan, keduanya memberikan hasil yang sama persis.

Setelah imputasi selesai dilakukan, data sekarang siap untuk digunakan.
